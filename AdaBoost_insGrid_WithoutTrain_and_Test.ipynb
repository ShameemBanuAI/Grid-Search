```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset.head()
```

```python
dataset.columns
```

```python
X=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
Y=dataset[['charges']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostRegressor
param_grid={ 'n_estimators':[25,50,100,1000],
           'loss': ['linear', 'square', 'exponential'],
           'random_state':[3,4,5]}
```

```python
grid=GridSearchCV(AdaBoostRegressor(),param_grid,verbose=3,n_jobs=-1)
grid.fit(X,Y)
```

```python
re=grid.cv_results_
bestParams=grid.best_params_
rscore=grid.best_score_
print("Hyper tuning parameter for  best r2score {} ".format(bestParams))
print("R2 Score value {} ".format(rscore))
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
age=int(input("Age : "))
bmi=int(input("Bmi : "))
children=int(input("Children : "))
male=int(input("Sex_Male 0 or 1 : "))
smoker=int(input("Smoker_Yes 0 or 1 : "))
```

```python
futurePred=grid.predict([[age,bmi,children,male,smoker]])
print("Future Predictions {} " .format(futurePred))
```
